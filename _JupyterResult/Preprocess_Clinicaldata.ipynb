```python
import pandas as pd
import numpy as np
```

```python
def CleaningData(d):
    
    df = d.copy()
    df['Gender'] = df['Gender'].astype('category')
    df['Gender'].cat.categories = [-1,1]
    df['Gender'] = df['Gender'].astype('float')
    
    df['ECOG PS'] = df['ECOG PS'].where(df['Hx Smoking'] != 'unknown' )
    df['ECOG PS'] = df['ECOG PS'].astype('category')
    df['ECOG PS'].cat.categories = [1,2,3,4,5, 0]
    df['ECOG PS'] = df['ECOG PS'].astype('float')
    
    
    df['Hx Smoking'] = df['Hx Smoking'].astype('category')
    df['Hx Smoking'].cat.categories = [3, 2, 1]
    df['Hx Smoking'] = df['Hx Smoking'].astype('float')
    #df['Hx Drinking'] = df['Hx Drinking'].where(df['Hx Drinking'] != 'unknown' )
    
    df['Hx Drinking'] = df['Hx Drinking'].astype('category')
    df['Hx Drinking'].cat.categories = [4, 5, 2, 3, 1, 0]
    df['Hx Drinking'] = df['Hx Drinking'].astype('float')
    
    df['T'] = df['T'].astype('category')
    df['T'].cat.categories = [1,2,3,4,5]
    df['T'] = df['T'].astype('float')
    
    df['N'] = df['N'].astype('category')
    df['N'].cat.categories = [1,2,3,4,5,6]
    df['N'] = df['N'].astype('float')
    
 
    
    df['Stage'] = df['Stage'].astype('category')
    df['Stage'].cat.categories = [4, 5, 2, 3, 1]
    df['Stage'] = df['Stage'].astype('float')
    
    df['p16'] = df['p16'].astype('category')
    df['p16'].cat.categories = [-10,10,0]
    df['p16'] = df['p16'].astype('float')
 
    df = pd.get_dummies(df, prefix='Sunsite_',  columns=['Subsite'])
    
    return df
```

```python
df1 = pd.read_csv("OPC/clinical_info_raw.csv")
feature = pd.read_csv("OPC/radiomic_features_name.csv")
```

```python
df = df1[['ID','Age','Gender','ECOG PS', 'Smoking PY', 'Hx Smoking', 'Hx Drinking', 'Subsite', 'T', 'N','Stage', 'p16']].copy()

```

```python
f = feature.T
f.reset_index(inplace = True)
f = f.merge(df, left_on = 'index' , right_on = 'ID')
```

```python
clinic = f[['ID','Age','Gender','ECOG PS', 'Smoking PY', 'Hx Smoking', 'Hx Drinking', 'Subsite', 'T', 'N','Stage', 'p16',26 ]].copy()
```

```python
clinic = CleaningData(clinic)
```

```python
c = clinic.set_index('ID').T
c = c.reset_index()
c.drop('index', axis = 1, inplace = True)
c.to_csv("OPC/VolumeAndClinic.csv")
```
